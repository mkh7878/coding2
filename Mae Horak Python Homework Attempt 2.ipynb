```python
# Load the image
img = cv2.imread('000004.jpg')

# Convert the image to RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get the red channel
red_channel = img[:, :, 0]

# Get the width of the image
width = img.shape[1]

# Find the highest red value in each column
highest_red = np.max(red_channel, axis=0)

# Sort the columns based on the highest red value
sorted_indices = np.argsort(highest_red)[::-1]  # reverse the order so that it's in descending order

# Rearrange the columns based on the sorted indices
sorted_img = np.zeros_like(img)
for i in range(width):
    sorted_img[:, i, :] = img[:, sorted_indices[i], :]

# Display the resulting image using imshow
plt.imshow(sorted_img)
plt.show()
```

```python
# Load the image
img = cv2.imread('000004.jpg')

# Convert the image to RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get the width of the image
width = img.shape[1]

# Rearrange the pixels in each column based on their RGB values
sorted_img = np.zeros_like(img)
for i in range(width):
    col = img[:, i, :]
    sorted_col = col[np.lexsort((col[:, 2], col[:, 1], col[:, 0]))]
    sorted_img[:, i, :] = sorted_col

# Display the resulting image using imshow
plt.imshow(sorted_img)
plt.show()
```
